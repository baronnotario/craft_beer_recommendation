```python
from itertools import izip, tee
```

```python
def pairwise(iterable):
    "s -> (s0,s1), (s1,s2), (s2, s3), ..."
    a, b = tee(iterable)
    next(b, None)
    return izip(a, b)
```

```python
diff = [y-x for x,y in pairwise(index)]
diff
```

```python
wrong_rows = []
for i in diff:
    if i != 13:
        wrong_rows.append (i)
print wrong_rows
```

```python
w=[]
for i in diff:
    if i != 13:
        w.append (index[index[i]])
print w
```
